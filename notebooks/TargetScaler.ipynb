```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler
import plotly.express as px
```

```python
up_threshold = 1
down_threshold = -1
```

```python
file = "../../target_1701887588.csv"
df = pd.read_csv(file)
df
```

```python
df['ss-trend'] = np.where(df['scaled_slope'] > up_threshold, 1,
                        np.where(df['scaled_slope'] < down_threshold, -1, 0))

df['ss-trend'].value_counts()
```

```python
scaled_slope = StandardScaler().fit_transform(df['slope'].values.reshape(-1, 1)).reshape(-1)
df['ss_slope'] = scaled_slope

df['ss-trend'] = np.where(df['ss_slope'] > up_threshold, 1,
                        np.where(df['ss_slope'] < down_threshold, -1, 0))

display(df['ss-trend'].value_counts())
display(df['ss_slope'].describe())

px.histogram(scaled_slope, nbins=1000)
```

```python
scaled_slope = RobustScaler().fit_transform(df['slope'].values.reshape(-1, 1)).reshape(-1)
df['rs_slope'] = scaled_slope

df['rs-trend'] = np.where(df['rs_slope'] > up_threshold, 1,
                        np.where(df['rs_slope'] < down_threshold, -1, 0))

display(df['rs-trend'].value_counts())
display(df['rs_slope'].describe())

px.histogram(scaled_slope, nbins=1000)
```
